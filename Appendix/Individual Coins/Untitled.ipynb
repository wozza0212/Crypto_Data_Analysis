```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
xrp = pd.read_csv('XRP.csv')
bitcoin = pd.read_csv('bitcoin.csv')
litecoin = pd.read_csv('litecoin.csv')
ethereum = pd.read_csv('ethereum.csv')
monero = pd.read_csv('monero.csv')
cardano = pd.read_csv('cardano.csv')
print(xrp.head())
```

```python
pop_coins = [xrp, bitcoin, litecoin, ethereum, monero, cardano]
for x in pop_coins:
    x['Formatted date'] = pd.to_datetime(x['Formatted date'])
print(bitcoin.info())
print(bitcoin.head())
```

```python
plt.plot(bitcoin['Formatted date'], bitcoin['Price($)'])
plt.show()
plt.close()
```

```python
for x in pop_coins:
    plt.plot(x['Formatted date'], x['Price($)'])
    plt.legend(x['Name'])
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()
    
```
